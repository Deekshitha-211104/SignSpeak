```python
import csv
import os
import json

# Path to JSON file
json_path = r"C:\Users\deeks_w4ub1k8\Sign2Speech\WLASL_v0.3.json"

# Path to folder where .mp4 videos are stored
video_folder = r"C:\Users\deeks_w4ub1k8\Sign2Speech\videos"

# Output CSV path
csv_output_path = r"C:\Users\deeks_w4ub1k8\Sign2Speech\video_gloss_mapping.csv"

# Load JSON
with open(json_path, "r") as f:
    wlasl_data = json.load(f)

# Build mapping: video_id → gloss
video_to_gloss = {}
for entry in wlasl_data:
    gloss = entry["gloss"]
    for instance in entry["instances"]:
        video_id = instance["video_id"]
        video_to_gloss[video_id] = gloss

# Create CSV file
with open(csv_output_path, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["video_path", "label"])  # header
    
    for filename in os.listdir(video_folder):
        if filename.endswith(".mp4"):
            video_id = os.path.splitext(filename)[0]
            if video_id in video_to_gloss:
                full_path = os.path.join(video_folder, filename)
                writer.writerow([full_path.replace("\\", "/"), video_to_gloss[video_id]])

print(f"CSV file saved at: {csv_output_path}")
```

```python
import pandas as pd

# Load CSV
df = pd.read_csv('video_gloss_mapping.csv')

# Load missing video IDs as strings with padding
with open('missing.txt', 'r') as file:
    missing_ids = set(line.strip().zfill(5) for line in file)

# Filter out rows where video_path contains a missing ID
def is_missing(path):
    video_id = path.split('/')[-1].split('.')[0]  # e.g., "07085"
    return video_id in missing_ids

df_cleaned = df[~df['video_path'].apply(is_missing)]

# Save cleaned version
df_cleaned.to_csv('video_gloss_mapping_cleaned.csv', index=False)

print(f"Removed {len(df) - len(df_cleaned)} missing entries.")
```
